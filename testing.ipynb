```python
import sqlite3
import pandas as pd

# 1. Create a connection to the database file
# Make sure 'loan_data.db' is in the same folder as your notebook
conn = sqlite3.connect('loan_data.db')

# 2. Write your SQL query
query = "SELECT * FROM loans"

# 3. Load into a DataFrame
df = pd.read_sql(query, conn)

# 4. Close the connection (Good practice!)
conn.close()

# 5. View the data
df.head()
```
